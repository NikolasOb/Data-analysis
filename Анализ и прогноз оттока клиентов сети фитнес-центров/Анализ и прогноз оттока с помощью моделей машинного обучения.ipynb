- 'Avg_additional_charges_total' — суммарная выручка от других услуг фитнес-центра: кафе, спорттовары, косметический и массажный салон.

### Выполним анализ распределения признаков по оттоку клиентов, построим гистограммы

```python
signs = df.drop('churn', axis=1)
fig, axes = plt.subplots(5,3, figsize=(15,20))
axes[4,1].set_axis_off()
axes[4,2].set_axis_off()

axes = axes.ravel()

for i, ax in zip(signs, axes):
    sns.histplot(df.query('churn==0')[i], ax=ax, label='Осталось', color='orange', alpha=0.3)
    sns.histplot(df.query('churn==1')[i], ax=ax, label='Ушло', color='blue', alpha=0.5)
    ax.set_ylabel('Частота')
    ax.set_title(f'Распределение признака {i}')
    
    ax.legend()
   
fig.tight_layout()

plt.show()
```

- Доля оставшихся для тех кто предпочитает 6 и 12 месячный абонемент существенно выше чем те кто предпочитает 1 месяц, оно и логично

## Построение модели прогнозирования оттока клиентов

### Разделим данные на обучающую и валидационную выборки, обучим модель и рассчитаем метрики

Как видели из матрицы корреляций есть две пары сильно коррелирующих признаков между собой, их корреляция более 0,8. В связи с мультиколлинеарностью данных признаков мы можем получить нерепрезентативные веса и переобученную под эти признаки модель. 
Удалим два признака их 4 так как по сути они действительно отражают одно и тоже и слишком очевидны.

```python
X_1 = df.drop(['churn', 'month_to_end_contract', 'avg_class_frequency_current_month'], axis=1) # удалим сильнокоррелирующие признаки и столбец churn
y = df['churn'] # заключим в переменную целевой признак
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_1, y, test_size=0.2, random_state = 0, stratify=y) # разделим выборку на обущающую и валидационную
scaler = StandardScaler() # стандартизируем и нормализуем данные 
scaler.fit(X_train)
X_train_st = scaler.transform(X_train)
X_test_st = scaler.transform(X_test)
```

```python
print(f'Размер тестовой выборки : {X_train.shape}')
print(f'Размер тестовой выборки : {X_test.shape}')
```

```python
models = [ # зададим список моделей
    LogisticRegression(random_state = 0, penalty='l2'),
    RandomForestClassifier(n_estimators = 100, random_state = 0),
]

# функция, которая принимает на вход модель и данные и выводит метрики
def make_prediction(m, X_train, y_train, X_test, y_test):
    model = m
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:,1]
    print(
        'Accuracy: {:.2f} / Precision: {:.2f} / Recall: {:.2f}'.format(
            accuracy_score(y_test, y_pred),
            precision_score(y_test, y_pred),
            recall_score(y_test, y_pred),
        )
    )


# напишем цикл, который выведет метрики по списку моделей
for i in models:
    print()
    print(f'Метрики для модели: {i}')
    make_prediction(m=i, X_train=X_train_st, y_train=y_train,
    X_test=X_test_st, y_test=y_test)
```

В целом обе модели показывают не плохие показатели метрик, но модель LogisticRegression оказалась более точна
